###### import necessary libraries

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
```

###### merge those two csv files (after getting as dataframes, get them as a single dataframe)

```python
college_1= pd.read_csv('college_1.csv')
college_2= pd.read_csv('college_2.csv')
merged= college_1.merge(college_2, how= 'outer')
print(merged)
```

###### Take each csv file , split that csv file into multiple categories (example csv files are added in the repo) 


###### consider if the codekata score exceeds 15000 points(present week) then make a csv on those observations as Exceeded expectations.csv


###### if  10000<codekata score<15000   (Reached_expectations.csv)



###### if  7000<codekata score<10000   (Needs_Improvement.csv)


###### if  codekate score < 7000        (Unsatisfactory.csv)

```python
Exceeded_expectations= merged.query("`CodeKata Score` > 15000")
Exceeded_expectations.to_csv('Exceeded_expectations.csv')
ex= pd.read_csv('Exceeded_expectations.csv')
print(ex)
```

```python
Reached_expectations= merged.query("`CodeKata Score`> 10000 and `CodeKata Score` < 15000")
Reached_expectations.to_csv("Reached_expectations.csv")
reach= pd.read_csv('Reached_expectations.csv')
print(reach)
```

```python
Needs_Improvement= merged.query("7000 < `CodeKata Score` < 10000")
Needs_Improvement.to_csv('Needs_Improvement.csv')
N_improve= pd.read_csv("Needs_Improvement.csv")
print(N_improve)
```

###### Average of previous week geekions vs this week geekions (i.e Previous Geekions vs CodeKata Score)

###### Department wise codekata performence (pie chart)

```python

data= merged.groupby(['Department'])['CodeKata Score'].max().to_frame()
data.rename(columns= {'Department': 'codekata_score'}, inplace=True)
data= data.reset_index()
# codekata= (data['codekata_score'])
# department= (data['Department'])
plt.pie(data['CodeKata Score'], labels= data['Department'])
plt.show()
```

###### Department wise toppers (horizantal bar graph or any visual representations of your choice)

```python
topper= merged.groupby(['Department'])['Name','python_en', 'CodeKata Score', 'python', 'mysql', 'Previous Geekions', 'computational_thinking'].max()
x= topper.reset_index()

sns.barplot(data= x, x='Name', y='python', hue='Department')

plt.show()
```
